```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt # to draw graphs
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/데이터마이닝수업/Data/Agent_Attrition.csv')
df.info()
df
```

```python
df.info()
df.columns
```

```python
# prompt: missing 값이 존재하는 'rank_INDEX14', 'rank_INDEX18',  'rank_INDEX23', 'rank_INDEX27', 'rank_INDEX28', 'rank_INDEX29', 'rank_INDEX32', 'rank_INDEX44', 'rank_INDEX93', 컬럼에 대해 최빈도값으로 대체하는 코드를 제공해주세요

# 'rank_INDEX14', 'rank_INDEX18',  'rank_INDEX23', 'rank_INDEX27', 'rank_INDEX28', 'rank_INDEX29', 'rank_INDEX32', 'rank_INDEX44', 'rank_INDEX93' 컬럼의 최빈값 계산
for col in ['rank_INDEX14', 'rank_INDEX18', 'rank_INDEX23', 'rank_INDEX27', 'rank_INDEX28', 'rank_INDEX29', 'rank_INDEX32', 'rank_INDEX44', 'rank_INDEX93']:
    mode_val = df[col].mode()[0]  # 최빈값
    df[col] = df[col].fillna(mode_val) # 최빈값으로 NaN 값 대체

# 확인 (선택적)
print(df[['rank_INDEX14', 'rank_INDEX18', 'rank_INDEX23', 'rank_INDEX27', 'rank_INDEX28', 'rank_INDEX29', 'rank_INDEX32', 'rank_INDEX44', 'rank_INDEX93']].info())
```

```python
# prompt: Gender, Q2, Q7, Q8, Q18, Q22, Q38 컬럼에 대해 문자형 타입으로 변환시키는 코드를 제공해주세요

# 'Gender', 'Q2', 'Q7', 'Q8', 'Q18', 'Q22', 'Q38' columns to string type
for col in ['Gender', 'Q2', 'Q7', 'Q8', 'Q18', 'Q22', 'Q38']:
    df[col] = df[col].astype(str)

df.info()
```

```python
# prompt: object 타입의 모든 변수를 one-hot enconder 함수를 이용하여 변화시키는 코드를 제공해주세요

from sklearn.preprocessing import OneHotEncoder

# object 타입의 컬럼 선택
object_cols = df.select_dtypes(include='object').columns

# OneHotEncoder 객체 생성
enc = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# object 타입 컬럼에 대해 one-hot encoding 적용
encoded_data = enc.fit_transform(df[object_cols])

# 인코딩된 데이터를 DataFrame으로 변환
encoded_df = pd.DataFrame(encoded_data, columns=enc.get_feature_names_out(object_cols))

# 원본 DataFrame에서 object 타입 컬럼 제거
df = df.drop(object_cols, axis=1)

# 원본 DataFrame에 인코딩된 데이터 추가
df = pd.concat([df, encoded_df], axis=1)

# 결과 확인
df.info()
#df.head
```

```python
df
```

```python
# prompt: df 데이터셋을 train dataset 70%, test dataset 30%로 분리하는 코드를 제공해주세요

from sklearn.model_selection import train_test_split

# 데이터프레임을 train 데이터셋과 test 데이터셋으로 분리 (70:30 비율)
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42) # random_state는 일정한 결과를 얻기위해 설정

# 분리된 데이터셋 확인 (선택 사항)
print("Train dataset shape:", train_df.shape)
print("Test dataset shape:", test_df.shape)
```

```python
# prompt: 종속변수를 Tgt_3YN 컬럼으로 하고 AgentID컬럼을 제외한 나머지를 설명변수해서 Logistic Regression 예측 모형을 이용하여 예측할 수 있는 코드를 제공해주세요

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# 설명 변수 (AgentID 제외)와 목표 변수 설정
X_train = train_df.drop(['AgentID', 'Tgt_3YN'], axis=1)
y_train = train_df['Tgt_3YN']
X_test = test_df.drop(['AgentID', 'Tgt_3YN'], axis=1)
y_test = test_df['Tgt_3YN']

# Logistic Regression 모델 생성 및 학습
model = LogisticRegression(max_iter=1000) # max_iter를 증가시켜 수렴 문제 해결 시도
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test)

# 모델 평가
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, y_pred))
```

```python
# prompt: 예측확률값에 대해  분포를 겹처서 확인 할 수 있는 코드를 제공해주세요

import matplotlib.pyplot as plt
import seaborn as sns

# 예측 확률 얻기
y_pred_proba = model.predict_proba(X_test)

# 클래스 0과 1에 대한 예측 확률 분포 시각화
plt.figure(figsize=(10, 6))

sns.kdeplot(y_pred_proba[y_test == 0, 1], label='Actual 0', shade=True)
sns.kdeplot(y_pred_proba[y_test == 1, 1], label='Actual 1', shade=True)

plt.xlabel('Predicted Probability of Class 1')
plt.ylabel('Density')
plt.title('Distribution of Predicted Probabilities by Actual Class')
plt.legend()
plt.show()
```

```python
# prompt: test 데이터셋 기준으로 예측값과 실제값을 비교할 수 있는 confusion matrix를 그릴 수 있는 코드를 제공해주세요

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion matrix 생성
cm = confusion_matrix(y_test, y_pred)

# Confusion matrix 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# prompt: 모델의 성능을 ROC Curve로 그릴 수 있는 코드를 제공해주세요

from sklearn.metrics import roc_curve, auc

# 예측 확률 얻기
y_pred_proba = model.predict_proba(X_test)[:, 1]

# ROC Curve 계산
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# AUC 계산
roc_auc = auc(fpr, tpr)

# ROC Curve 시각화
plt.figure(figsize=(10,6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='baseline')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

plt.legend()
plt.show()
```

```python
# prompt: 예측 확률값에 대한 분포를 확인할 수 있는 그래프를 그리는 코드를 제공해주세요

import matplotlib.pyplot as plt

# 예측 확률값에 대한 분포 확인 (히스토그램)
plt.figure(figsize=(10, 6))
plt.hist(y_pred_proba, bins=20, edgecolor='black')  # bins는 히스토그램의 구간 개수 조절
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.title('Distribution of Predicted Probabilities')
plt.show()

# 예측 확률값에 대한 분포 확인 (밀도 그래프)
plt.figure(figsize=(10, 6))
sns.kdeplot(y_pred_proba, shade=True)
plt.xlabel('Predicted Probability')
plt.ylabel('Density')
plt.title('Distribution of Predicted Probabilities')
plt.show()
```
