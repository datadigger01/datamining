### 2.3 숫자타입으로 지정된 Gender 컬럼을 문자타입으로 변환

```python
# prompt: rank_INDEX14', 'rank_INDEX18', 'rank_INDEX23', 'rank_INDEX27', 'rank_INDEX28', 'rank_INDEX29', 'rank_INDEX32', 'rank_INDEX44', 'rank_INDEX93', 'rank_AVG_AMT_6M' 컬럼에 대해 NaN missing 값을 최빈값으로 대체하고 그이후에 해당 컬럼에 대해 정수형 타입으로 변환하는 코드를 제공해주세요

columns_to_process = ['rank_INDEX14', 'rank_INDEX18', 'rank_INDEX23', 'rank_INDEX27', 'rank_INDEX28', 'rank_INDEX29', 'rank_INDEX32', 'rank_INDEX44', 'rank_INDEX93', 'rank_AVG_AMT_6M']

for column in columns_to_process:
  # 최빈값 찾기
  most_frequent_value = df[column].mode()[0]

  # NaN 값을 최빈값으로 대체
  df[column].fillna(most_frequent_value, inplace=True)

  # 정수형으로 변환
  df[column] = df[column].astype(int)

# 결과 확인
df.info()
```

```python
# Gender 컬럼을 문자타입으로 변환
df['Gender'] = df['Gender'].astype(str)

df.info()
```

# 3. 데이터 시각화(Visualization)를 통한 데이터 탐색(Exploration)

```python
# prompt: Tgt_3YN 컬럼을 제외한 나머지 모든 컬럼에 대해 정렬된 값별로 bar 그래프를  3열로그리는 코드를 제공해주세요

import matplotlib.pyplot as plt

# 'Tgt_3YN' 컬럼 제외한 나머지 컬럼 가져오기
columns_to_plot = [col for col in df.columns if col != 'Tgt_3YN']

# 3열로 그래프 그리기 위한 설정
num_cols = 3
num_rows = (len(columns_to_plot) + num_cols - 1) // num_cols

# 그래프 그리기
fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, 5 * num_rows))
fig.tight_layout(pad=5.0)

for i, column in enumerate(columns_to_plot):
  row = i // num_cols
  col = i % num_cols

  # 범주형 변수인 경우 value_counts()로 빈도수 계산
  if df[column].dtype == object:
    value_counts = df[column].value_counts()
    axes[row, col].bar(value_counts.index, value_counts.values, color='gray')
    axes[row, col].set_title(f'{column} (Value Counts)')
    axes[row, col].tick_params(axis='x', rotation=45, labelsize=8)

  # 숫자형 변수인 경우 정렬된 값으로 히스토그램 그리기
  else:
    df[column].value_counts().sort_index().plot(kind='bar', color='gray',ax=axes[row, col])
    axes[row, col].set_title(f'{column} (Sorted Values)')
    axes[row, col].tick_params(axis='x', rotation=45, labelsize=8)

  # 빈 공간 제거
  if i >= len(columns_to_plot):
    fig.delaxes(axes[row, col])

plt.show()
```

```python
# prompt: 'Age_Grp', 'Gender', 'Apply_CH', 're_hire_freq_grp' 컬럼에 대해서 왼쪽축 기준으로는 정렬된 컬럼값 기준의 빈도수를 오른쪽축기준으로는 Tgt_3YN 값에 대한 발생비율이 최대값이 0.6으로 고정된 그래프로 그리되, 3열로 보여지게 하는 코드를 제공해주세요. 그리고 막대그래프를  연한회색으로하고 선그래프는 파란색으로 셋팅해주세요

import matplotlib.pyplot as plt

# 컬럼 리스트
columns_to_plot = ['Age_Grp', 'Gender', 'Apply_CH', 're_hire_freq_grp','Q2',
       'Q7', 'Q8', 'Q18', 'Q22', 'Q38', 'rank_INDEX14', 'rank_INDEX18',
       'rank_INDEX23', 'rank_INDEX27', 'rank_INDEX28', 'rank_INDEX29',
       'rank_INDEX32', 'rank_INDEX44', 'rank_INDEX93', 'rank_AVG_AMT_6M']

# 3열로 그래프 그리기 위한 설정
num_cols = 3
num_rows = (len(columns_to_plot) + num_cols - 1) // num_cols

# 그래프 그리기
fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, 5 * num_rows))
fig.tight_layout(pad=5.0)

for i, column in enumerate(columns_to_plot):
  row = i // num_cols
  col = i % num_cols

  # 범주형 변수에 대해 빈도수 및 Tgt_3YN 비율 계산
  value_counts = df[column].value_counts().sort_index()
  tgt_3yn_ratio = df.groupby(column)['Tgt_3YN'].mean()

  # 왼쪽 축에 막대 그래프 그리기 (빈도수)
  axes[row, col].bar(value_counts.index, value_counts.values, color='lightgray')
  axes[row, col].set_xlabel(column)
  axes[row, col].set_ylabel('Frequency')

  # 오른쪽 축에 선 그래프 그리기 (Tgt_3YN 비율)
  axes2 = axes[row, col].twinx()
  axes2.plot(tgt_3yn_ratio.index, tgt_3yn_ratio.values, color='darkred')
  axes2.set_ylabel('Tgt_3YN Ratio')
  axes2.set_ylim(0, 0.6)  # y축 범위를 0에서 0.6으로 고정

  axes[row, col].set_title(f'{column} Frequency and Tgt_3YN Ratio')
  axes[row, col].tick_params(axis='x', rotation=45, labelsize=8)

  # 빈 공간 제거
  if i >= len(columns_to_plot):
    fig.delaxes(axes[row, col])


plt.show()

```

# 4. 변수변환
- 정규화/표준화
- 문자형 변수값을 숫자형태의 값으로 전환 ( 보통 마이닝 알고리즘에서는 숫자형 형태의 값만 인식 )

```python
df.info()
```

```python
# prompt: 문자형 컬럼만 선택해서 scikit learn 함수에서 제공하는 one-hot encoding 함수를 이용하여 새로운 컬럼을 생성하는 코드를 제공해주세요

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# 문자형 컬럼만 선택
categorical_columns = df.select_dtypes(include=['object']).columns

# OneHotEncoder 객체 생성
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# 문자형 컬럼에 대해 one-hot encoding 수행
encoded_data = encoder.fit_transform(df[categorical_columns])

# 새로운 컬럼 이름 생성
encoded_column_names = encoder.get_feature_names_out(categorical_columns)

# 새로운 컬럼을 DataFrame에 추가
encoded_df = pd.DataFrame(encoded_data, columns=encoded_column_names, index=df.index)

# 기존 DataFrame과 one-hot encoded DataFrame 병합
df_encoded = pd.concat([df, encoded_df], axis=1)

# 기존 문자형 컬럼 삭제
df_encoded = df_encoded.drop(columns=categorical_columns)

# 결과 확인
df_encoded.info()
df_encoded
```

```python
# df_encoded

# 'Tgt_3YN' 컬럼 제외한 나머지 컬럼 가져오기
columns_to_plot = [col for col in df_encoded.columns if col != 'Tgt_3YN']

# 3열로 그래프 그리기 위한 설정
num_cols = 3
num_rows = (len(columns_to_plot) + num_cols - 1) // num_cols

# 그래프 그리기
fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, 5 * num_rows))
fig.tight_layout(pad=5.0)

for i, column in enumerate(columns_to_plot):
  row = i // num_cols
  col = i % num_cols

  # 범주형 변수인 경우 value_counts()로 빈도수 계산
  if df_encoded[column].dtype == object:
    value_counts = df_encoded[column].value_counts()
    axes[row, col].bar(value_counts.index, value_counts.values, color='gray')
    axes[row, col].set_title(f'{column} (Value Counts)')
    axes[row, col].tick_params(axis='x', rotation=45, labelsize=8)

  # 숫자형 변수인 경우 정렬된 값으로 히스토그램 그리기
  else:
    df_encoded[column].value_counts().sort_index().plot(kind='bar', color='gray',ax=axes[row, col])
    axes[row, col].set_title(f'{column} (Sorted Values)')
    axes[row, col].tick_params(axis='x', rotation=45, labelsize=8)

  # 빈 공간 제거
  if i >= len(columns_to_plot):
    fig.delaxes(axes[row, col])

plt.show()
```
