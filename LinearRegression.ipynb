```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt # to draw graphs
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/데이터마이닝수업/Data/BostonHousing.csv')
df
```

+ CRIM: 범죄율
+ ZN: 25,000평방피트기준 거주지 비율
+ INDUS: 비소매업종 점유구역 비율
+ CHAS:찰스강 인접여부(인접=1, 비인접:0)
+ NOX:일산화 질소 농도
+ RM: 거주지의 평균 방 개수
+ AGE:1940년 이전에 건축된 주택에 사는 비율
+ DIS:보스톤 5대 상업지구와의 거리
+ RAD:고속도로 진입용이성 정도
+ TAX: 재산세율(10,000달러당)
+ PTRATIO: 학생 대 교사 비율
+ LSTAT: 저소득층 비율
+ MEDV : 주택가격의 중앙값(단위:1,000달러)
+ CAT.MEDV: 주택가격의 중앙값이 3만달러 이상여부(이상=1,이하=0)

```python
# predicators & target
features = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'LSTAT']
label = ['MEDV']
```

## LinearRegression 모듈 호출

```python
import statsmodels.api as sm
```

```python
model = sm.OLS(df[label], df[features]).fit()
```

```python
print(model.summary())
```

```python
from sklearn.model_selection import train_test_split
# Split Data for Logistic Regression
X_train, X_test, y_train, y_test = train_test_split(df[features], df[label], test_size=0.20, random_state=365)
X_train
#y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm_model=LinearRegression().fit(X_train, y_train)
```

```python
df_colnm_lm = pd.DataFrame(lm_model.feature_names_in_, columns=['feature'])
df_coef_lm = pd.DataFrame(lm_model.coef_.T, columns=['coeffiecent'])

coef_tb = pd.concat([df_colnm_lm, df_coef_lm], axis=1)
coef_tb
```

```python
pred_value = lm_model.predict(X_test)
```

```python
lm_pred_value = pd.DataFrame(pred_value, columns=['lm_predicted_value'])

real_value = y_test.reset_index()


rslt=pd.concat([real_value, lm_pred_value],axis=1)
rslt
```

```python
plt.figure(figsize=(18,6))

# 예측값 Plot
plt.plot(rslt['lm_predicted_value'], marker='o',markersize=6,color='red', linestyle='dashed', linewidth=1,label='lasso_forecasted_value')

# 예측된 결과값 Plot
plt.plot(rslt['MEDV'], marker='o',markersize=6,color='blue',linewidth=1,label='real_value:MEDV' )

plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, mean_absolute_percentage_error


print("lm_MAE  :",mean_absolute_error(pred_value, y_test))

print("================================")
print("_MAPE  :",mean_absolute_percentage_error(pred_value, y_test))

print("================================")
print("lm_RMSE  :",mean_squared_error(pred_value, y_test)**0.5)

print("================================")
print("lm_MAPE  :",mean_absolute_percentage_error(pred_value, y_test))
```
