# 데이터시각화(Visualization)

+ Histogram

```python
#plt.hist(df.CRIM)
plt.hist(df.CRIM, bins=20)
#plt.hist(df.CRIM, bins=20, density=False, rwidth=0.9)

#plt.grid(axis='y', alpha=0.75)
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Histgram')
plt.show()
```

+ plotting

```python
plt.plot(df.MEDV, marker='o')

#plt.xlabel('X')
plt.ylabel('HousePrice')
plt.title('Plotting')
plt.show()
```

+ 산점도

```python
plt.scatter(df.CRIM, df.MEDV)

plt.xlabel('CRIM')
plt.ylabel('HousePrice')
plt.title('Scatter')
plt.show()
```

+ BoxPlot

```python
plt.boxplot([df.MEDV])
plt.show()
```

```python
sns.boxplot(y="MEDV", x="CAT. MEDV", data=df)
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(12,6))

ax[0,0].hist(df.CRIM)
ax[0,1].plot(df.MEDV)
ax[1,0].scatter(df.CRIM, df.MEDV)
ax[1,1].boxplot(df.MEDV)

plt.show()
```
